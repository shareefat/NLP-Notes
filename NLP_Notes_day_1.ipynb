#TF-IDF
the first term is term frequency (TF) the number of times a word appeares in a document divided by total number of word in that document

the second term is the inverse document frequency computed as the logaritham of the number of the document in the corpus divided by the number of document where the specific term apperes

steps for performing TF-IDF

1. Tokenization
2. TF -IDF calculating
3. Construct the metrix
4. Normalization

#WORD EMBEDDING IN NLP


word embedding is an approch for representing word and documents . word embedding or word vector is a numeric vector input that represent a word in a lower diamensional space it allow words with similar meaning to have similar representation


Need for Word Embedding;

* To reduce diamensionality

* to use a word to predict the word around it

* relation b/w the word

How are word Embedding Used.

* They are used as input to machine learning model

* take the word --> give the numeric representation ---> use in training or inference

* to represent or visualize any underlying pattern of usage in the corpus that as used for training
